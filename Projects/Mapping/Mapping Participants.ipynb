## Working with Postcodes

Most of us will will not know the longatude and latitudes of a given location, but we may well know the postcode, for this reason its more common to capture postcodes than any other form of location data.  

Postcode data however is not terribly useful in its own right, postcode areas highly irregular, different postcodes are different sizes and postcode sysyem is not layed out in a any form of regular pattern. It is therefore very difficult to get insights into geographical groupings from post code alone.

## Getting Co-ordinates

Before we can start plotting locations it is therefore necessary to be able to convert postcode into some form of co-ordinate system. There are several co-ordinate systems to choose. For example longatude and latitudes or grid reference.

There are many ways that you can get x and y co-ordinates from postcodes. I used a [freemaptools](https://www.freemaptools.com/convert-uk-postcode-to-lat-lng.htm)

If you wanted you could automate this process by using google's [geocoding api](https://developers.google.com/maps/documentation/geocoding/intro#geocoding) for developers.

# Mapping Venues with plot with matplotlib

The Earth is of course a sphere, and if your datapaoints are further apart or you want to plot them in more accuracy you would need to adjust for this. # Mapping Venues with Folium

While a scatter plot gives you a very basic idea of the location of the datapoints in relation to each other, if you want to see how that relates to features like transport networks, facilities you really need to plot this on a map.

![Folium Map](Images/foliummap.png?raw=true)

[Folium](https://github.com/python-visualization/folium) is a library that allows you to plot maps in python. In this example i've created a map with Folium, set the default view and passed the same longitude and latitude values and venue names and as markers to our map.

```python
# Create a map using the folium module

# import libraries
import folium
import pandas as pd
import pylab

# import the data into a dataframe
df=pd.read_csv('Data/baltic_venues.csv', sep=',')

# set default position and scale
lon, lat = -1.49828,   54.79921
zoom_start = 10

# initiate map
venue_map = folium.Map(
    location=[lat, lon],
    tiles='OpenStreetMap',
    zoom_start=zoom_start,
    control_scale=True)

# create datapoints with loop
allvenues=[]
for i, row in df.iterrows():
    venue={'loc': [row['Latitude'], row['Longitude']], 'label':row['Venue']}
    allvenues.append(venue)

# Add markers to the map
for venue in allvenues:
    marker = folium.Marker(location=venue['loc'], popup=venue['label'])
    marker.add_to(venue_map)

# The last object in the cell always gets shown in the notebook
pylab.savefig('Maps/folium.png')
venue_map.save('Maps/foliumscale.html')
venue_map
```

# Travel Distance

Because we have the location of the participants and the venues we can very crudely calculate travel distances. You have to do a little maths to compensate for the curvature of the earth and the distance are the direct distance between the 2 points not taking into account the means of travel, obstructions or route choices.

```python
# import modules
import pandas as pd
from math import sin, cos, sqrt, atan2, radians

# import participant data
Activitiesfile="Data/All_Activities.csv"
Participants=pd.read_csv(Activitiesfile, encoding = "ISO-8859-1", sep=',')

# Calculate Individula Travel distance and add to distance column
R = 3956 # Radius of earth use 6373.0 for kilometers or 3956 for miles

for index, row in Participants.iterrows():
    dlon =(radians(float(row["venue long"]))-(radians(float(row["attend Longitude"]))))
    dlat =((radians(float(row["venue lat"])))-(radians(float(row["attend Latitude"]))))
    a = sin(dlat / 2)**2 + cos(row["venue lat"]) * cos(row["attend Latitude"]) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    Participants.set_value(index, "Distance", distance, takeable=False)

# Calculate Average Travel distances

print("-------------------------------------")
print("Distance Travelled")
print("-------------------------------------")
print("The shortest distance travelled to attend an activity was " + (str(min((Participants)["Distance"]))) + " miles")
print("The largest distance travelled to attend an activity was " + (str(max((Participants)["Distance"]))) + " miles")
print("The average (mean) distance travelled to attend an activity was " + (str(float((Participants)["Distance"].mean()))) + " miles")
print("The average (median) distance travelled to attend an activity was " + (str(float((Participants)["Distance"].median()))) + " miles")
# print("The average (mode) distance travelled to attend an activity was " + (str(float((Participants)["Distance"].mode()))) + " miles")
print("-------------------------------------")  
```

```python
# import modules
import pandas as pd
import numpy as np
from math import sin, cos, sqrt, atan2, radians
import matplotlib.pyplot as plt
from matplotlib import cm
import pylab
%matplotlib inline

# import all participant data & replace blank Numbers with default of 1
Activitiesfile="Data/All_Activities.csv"
Participants=pd.read_csv(Activitiesfile, encoding = "ISO-8859-1", sep=',')
for column in Participants:
    Participants['Number'] = Participants['Number'].replace(r'\s+', np.nan, regex=True)
    Participants['Number'] = Participants['Number'].fillna(1) 

# initiate empty dictionaries
Period_dict={}   # for initial filtering
Venue_dict={}; Gates_venue_dict={}; Gates_venues_dict={}
Session_dict={}; Gates_session_dict={};Session_count_dict={}
Instance_dict={}; Gates_instance_dict={}
Venue_act_dict={}; Gates_venue_act_dict={}
Activity_dict={}; Gates_activity_dict={}
Season_dict={}; Gates_season_dict={}
Date_dict={}; Gates_date_dict={};Colour_date_dict={}

# get and set filtering input 
print("This project contains data from :")  
for index, row in Participants.iterrows():
    if row['Season'] not in Period_dict.keys():
        Period_dict[row['Season']]=1
print(Period_dict.keys())        
default = "all"
Selected_season = input("Please enter Period name:")
Selected_season=str(Selected_season)
if Selected_season in Period_dict.keys():
    Participants = Participants[Participants['Season'] == Selected_season]
    
# process time/date data
Participants['Dateformat'] = pd.to_datetime(Participants['Date'], format="%d/%m/%y")    
    
#create a dictionary of gateshead LA postcodes   
postcodefile="Data/Gatesheadpostcodes.csv"
postcodes=pd.read_csv(postcodefile)
postcodedict = {}

for c in postcodes["postcode"]:
    if c not in postcodedict.keys():
        postcodedict[c]=1 
    else:
        postcodedict[c]=((postcodedict)[c])+1
        
# initiate empty dictionaries
Venue_dict={}; Gates_venue_dict={}; Gates_venues_dict={}
Session_dict={}; Gates_session_dict={};Session_count_dict={}
Instance_dict={}; Gates_instance_dict={}
Venue_act_dict={}; Gates_venue_act_dict={}
Activity_dict={}; Gates_activity_dict={}
Season_dict={}; Gates_season_dict={}
Date_dict={}; Gates_date_dict={};Colour_date_dict={}

# establish counts
Household_count=0; Gates_household_count=0
Participant_count=0; Gates_participant_count=0

# Put Participant data into relevant dictionaries and increment counts
for index, row in Participants.iterrows():
    Household_count=Household_count+1
    Participant_count=Participant_count+(int(row['Number']))
    
    #Venue
    if row['Venue'] in Venue_dict.keys():
        Venue_dict[row['Venue']]=((Venue_dict[row['Venue']])+(int(row['Number'])))
    else:
        Venue_dict[row['Venue']]=(int(row['Number']))
        
    #Session
    if row['Session'] in Session_dict.keys():
        Session_dict[row['Session']]=((Session_dict[row['Session']])+(int(row['Number'])))
    else:
        Session_dict[row['Session']]=(int(row['Number']))  
        
    #Instance
    Instance=row['Venue']+"_"+row['Session']+"_"+row['Date']
    if Instance in Instance_dict.keys():
        Instance_dict[Instance]=Instance_dict[Instance]+(int(row['Number'])) 
        if row['Session'] in Activity_dict.keys():
            Activity_dict[row['Session']]=Activity_dict[row['Session']]+1
        else:
            Activity_dict[row['Session']]=1    
    else:
        Instance_dict[Instance]=(int(row['Number']))  
        if row['Session'] in Session_count_dict.keys():
            Session_count_dict[row['Session']]=Session_count_dict[row['Session']]+1
        else:
            Session_count_dict[row['Session']]=1 
        if row['Venue'] in Venue_act_dict.keys():
            Venue_act_dict[row['Venue']]=Venue_act_dict[row['Venue']]+1
        else:
            Venue_act_dict[row['Venue']]=1
            
    #Season
    if row['Season'] in Season_dict.keys():
        Season_dict[row['Season']]=((Season_dict[row['Season']])+(int(row['Number'])))
    else:
        Season_dict[row['Season']]=(int(row['Number'])) 
        
    #Date
    if row['Dateformat'] in Date_dict.keys():
        Date_dict[row['Dateformat']]=((Date_dict[row['Dateformat']])+(int(row['Number'])))
    else:
        Date_dict[row['Dateformat']]=(int(row['Number']))
        Colour_date_dict[row['Dateformat']]=(str(row['Venue']))
       
# Put Gateshead Participant data into relevant dictionaries and increment counts

for index, row in (Participants[Participants['Postcode'].isin(postcodedict)]).iterrows():
    Gates_household_count=Gates_household_count+1
    Gates_participant_count=Gates_participant_count+(int(row['Number']))
    
    #Venue
    if row['Venue'] in Gates_venue_dict.keys():
        Gates_venue_dict[row['Venue']]=((Gates_venue_dict[row['Venue']])+(int(row['Number'])))
    else:
        Gates_venue_dict[row['Venue']]=(int(row['Number']))
    #Session
    if row['Session'] in Gates_session_dict.keys():
        Gates_session_dict[row['Session']]=((Gates_session_dict[row['Session']])+(int(row['Number'])))
    else:
        Gates_session_dict[row['Session']]=(int(row['Number']))  
        
    #Instance
    Instance=row['Venue']+"_"+row['Session']+"_"+row['Date']
    if Instance in Gates_instance_dict.keys():
        Gates_instance_dict[Instance]=Gates_instance_dict[Instance]+(int(row['Number'])) 
        if row['Venue'] in Gates_venue_act_dict.keys():
            Gates_venue_act_dict[row['Venue']]=Gates_venue_act_dict[row['Venue']]+1
        else:
            Gates_venue_act_dict[row['Venue']]=1
        if row['Session'] in Gates_activity_dict.keys():
            Gates_activity_dict[row['Session']]=Gates_activity_dict[row['Session']]+1
        else:
            Gates_activity_dict[row['Session']]=1            
    else:
        Gates_instance_dict[Instance]=(int(row['Number']))  
        
    #Season
    if row['Season'] in Gates_season_dict.keys():
        Gates_season_dict[row['Season']]=((Gates_season_dict[row['Season']])+(int(row['Number'])))
    else:
        Gates_season_dict[row['Season']]=(int(row['Number']))
    #Date 
    if row['Dateformat'] in Gates_date_dict.keys():
        Gates_date_dict[row['Dateformat']]=((Gates_date_dict[row['Dateformat']])+(int(row['Number'])))
    else:
        Gates_date_dict[row['Dateformat']]=(int(row['Number']))
        Gates_date_dict[row['Dateformat']]=(int(row['Number']))
        
# Put Gateshead venues into a dictionary and increment counts

Gates_venues_dict={}
for index, row in (Participants[Participants['Venue Postcode'].isin(postcodedict)]).iterrows():
    if row['Venue'] not in Gates_venues_dict.keys():
        Gates_venues_dict[row['Venue']]=1
        
# Ensure all venues are in both dictionaries
for Venue in Venue_dict:
    if Venue not in Gates_venue_dict:
        Gates_venue_dict[Venue]=0  
        
# Ensure all session are in both dictionaries
for session in Session_dict:
    if session not in Gates_session_dict:
        Gates_session_dict[session]=0       
        
# Ensure all Activities are in both dictionaries
for activity in Activity_dict:
    if activity not in Gates_activity_dict:
        Gates_activity_dict[activity]=0 
        
# Ensure all Dates are in both dictionaries (set to 1 if doesn't exist for scaling)
for activity in Activity_dict:
    if activity not in Gates_activity_dict:
        Gates_activity_dict[activity]=0        
         
# Display the Summary info 
print("\n-------------------------------------")
print("Participant Infomation:")
print("-------------------------------------\n")
print((str(Participant_count)) + " participants, with " + str(len(np.unique(list(Participants['Postcode'])))) + " unique postcodes, took part in " + str(len(Instance_dict.keys()))+ " activities ")
print((str(Gates_participant_count)) + " of those participants came from within the Gateshead Local Authority Area")
# print("-------------------------------------\n")        
   
# Graph Participation by date
Date_list=(list(sorted(Date_dict.keys())))
Participation_list = [Date_dict[key] for key in sorted(Date_dict.keys(), reverse=False)]
scale_list =[Gates_date_dict[key] for key in sorted(Gates_date_dict.keys(), reverse=False)]
scaler=(400/(max(Gates_date_dict.values())))
for idx, i in enumerate(scale_list):
    scale_list[idx] = float(i) * float(scaler)

# Set Colours for venues   
col=0
Colour_dict={}
Colour_list=("yellow","orange","sienna","tan","red","purple","navy","blue", "cyan", "green","green","violet","grey","black","pink","brown","slate blue")
for idx,v in Colour_date_dict.items():
    if v not in Colour_dict:
        Colour_dict[v]=Colour_list[col]
        col=col+1               
# Colour_dict = {v: k for k, v in Colour_dict.items()}

for k,v in Colour_date_dict.items():
    if v in Colour_dict : Colour_date_dict[k]=Colour_dict[v]
                
Venue_colour_list = [Colour_date_dict[key] for key in sorted(Date_dict.keys(), reverse=False)]

plt.figure(1, figsize=(8, 6))                
plt.scatter(Date_list,Participation_list,alpha=0.4, s=scale_list,c=Venue_colour_list)
plt.xlabel("Date")
plt.ylabel("Participation")
plt.title('Participation by Date');

# limit the axis to include only what you want to display
plt.xlim(['2017-03-15','2017-10-01'])
plt.xticks(rotation='90')

# print("\n Conversions by time and date : \n ")
pylab.savefig('Graphs/Participation.png')   # save the figure to file
pylab.savefig('Graphs/Participation.pdf') # savefig('salesdistribution.pdf')
plt.show()      

# Display the Activity info       
print("\n-------------------------------------")
print("Activity Infomation:")
print("-------------------------------------\n")

print("There were a total of " + str(len(Instance_dict.keys())) + " events")
print("At " + str(len(Venue_dict.keys()))  + " locations including " + str(len(Session_dict))+ " categories of activity\n")
# print(Instance_dict.keys())  

# Graph the Actitity type data
plt.figure(3,figsize=(8, 6))
SValues=(list(Session_dict.values()))
GSValues=(list(Gates_session_dict.values()))
labels=(list(Session_dict.keys()))
x = list(range(len(Session_dict.values())))
plt.bar(x,SValues) # barchart
plt.bar(x,GSValues) # barchart
# plt.plot(SValues) # line graph
plt.xticks(x, labels,rotation='vertical')
plt.ylabel('Participants')
plt.xlabel('Type of Activity')

# Print the Activity
pylab.savefig('Graphs/Activity.png',bbox_inches = 'tight')   # save the figure to file
pylab.savefig('Graphs/Activity.pdf') # savefig('salesdistribution.pdf')
plt.show()  
    
print("\n-------------------------------------")
```
